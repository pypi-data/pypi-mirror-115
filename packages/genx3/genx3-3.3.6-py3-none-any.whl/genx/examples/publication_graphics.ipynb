# Simple example to generate publication graphs using GenX API
Load a model that you want to plot

```python
model,optimizer=api.load('SuperAdam_SiO_full_res_calc.hgx')
```

## Graphs using the line settings from the GenX model

```python
figure(figsize=(8,6)) #  size of figure in inches
model.plot(data_labels=['data'], sim_labels=['model'])
xlabel(r'q ($\AA^{-1}$)')
ylabel('reflecitivty (a.u.)')
#savefig('plot_example.png', dpi=600) #  export the graphics, see matplotlib savefig for export options
```

## Graph from scratch and manually plot data and simulation

```python
model.simulate() #  in case the model was not simulated before (automatically done in model.plot) y_sim is missing
figure(figsize=(14,4)) #  size of figure in inches

subplot(121)
ds=model.data[0]
errorbar(ds.x, ds.y, yerr=ds.error, color='#00aa00', label='data')
semilogy(ds.x, ds.y_sim, color='#000000', label='model')
legend()
title('overview')

xlabel(r'q ($\AA^{-1}$)')
ylabel('reflecitivty (a.u.)')

subplot(122)
ds=model.data[0]
errorbar(ds.x, ds.y, yerr=ds.error, color='#00aa00', label='data', fmt='o')
semilogy(ds.x, ds.y_sim, color='#000000', label='model')
legend()
xlim(0.01, 0.025)
ylim(1e-2, 1)
title('zoom')

xlabel(r'q ($\AA^{-1}$)')
ylabel('reflecitivty (a.u.)')
#savefig('plot_example2.png', dpi=600) #  export the graphics, see matplotlib savefig for export options
```
